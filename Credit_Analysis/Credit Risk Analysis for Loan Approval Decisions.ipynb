```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)

n = 800

df = pd.DataFrame({
    "Income": np.random.normal(60000,15000,n),
    "Credit_Score": np.random.normal(650,70,n),
    "Loan_Amount": np.random.normal(20000,8000,n),
    "Age": np.random.randint(21,65,n),
})

df["Default"] = (
    (df["Credit_Score"] < 600) |
    (df["Income"] < 40000) &
    (df["Loan_Amount"] > 25000)
).astype(int)

df.head()
```

```python
df["Default"].value_counts().plot(kind="bar")

plt.title("Loan Default Distribution")
plt.xlabel("Default Status (0 = No Default, 1 = Default)")
plt.ylabel("Number of Borrowers")

plt.xticks(rotation=0)

plt.show()
```

Insight:

Loan defaults occur less frequently but represent high financial risk.

```python
plt.scatter(df["Credit_Score"], df["Default"], alpha=0.3)

plt.xlabel("Credit Score")
plt.ylabel("Default (0 = No Default, 1 = Default)")

plt.title("Credit Score vs Loan Default")

plt.show()
```

Insight:

The scatter plot shows borrower credit scores against loan default status. Borrowers with lower credit scores appear more frequently in the default category, suggesting that credit score is an important factor in predicting loan default risk.

```python
plt.scatter(df["Income"], df["Loan_Amount"], alpha=0.4)

plt.xlabel("Income")
plt.ylabel("Loan Amount")

plt.title("Income vs Loan Amount")

plt.show()
```

Insight:

The scatter plot shows the relationship between borrower income and loan amount. The distribution appears widely scattered, indicating a weak correlation between these variables in the simulated dataset. In real-world financial datasets, lenders typically consider income when determining loan eligibility, which would likely produce a stronger relationship.

```python
df["Loan_to_Income"] = df["Loan_Amount"] / df["Income"]
```

```python
plt.hist(df["Loan_to_Income"], bins=30)

plt.title("Loan-to-Income Ratio Distribution")
plt.xlabel("Loan to Income Ratio")
plt.ylabel("Frequency")

plt.show()
```

Insight:

Borrowers with higher loan-to-income ratios may represent higher credit risk, as their loan obligations are large relative to their income.

```python
sns.heatmap(df.corr(),annot=True)

plt.title("Credit Risk Correlation")

plt.show()
```

Conclusion

Explains:

Credit score strongly affects default probability

Loan-to-income ratio is a key risk factor

Data analytics helps banks assess borrower risk
